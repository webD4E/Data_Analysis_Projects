````python
# to make all the values of cell into float 

def mk_float(df):
    # to iterate through columns
    for clm in df.columns:
        if clm == "Country":
            continue
        df[clm] = pd.to_numeric(df[clm], errors="coerce")
    return df

df1 = mk_float(df1)
df1.head()
````
